# Functions

```python
#Gives y
def pred(w,x):
  n=len(x)
  y=[]
  for i in x:
    v=1
    d=0
    sum=0
    while(d<len(w)):
      sum=sum+v*w[d]
      v=v*i
      d=d+1
    y.append(sum)
  y=np.array(y)
  return y
```

```python
#Finds the parameters
def param(x,t,lam,d):
  n=len(x)
  a=np.zeros((d+1,d+1))
  b=[]
  for j in range (0,d+1):
    for i in range (0,d+1):
      val=0
      for k in range (0,n):
        val=val+x[k]**(i+j)
      if i==j:
        val=val+lam
      a[j][i]=val

  for j in range (0,d+1):
    val=0
    for k in range (0,n):
      val=val+t[k]*(x[k]**j)
    b.append(val)
  w=np.linalg.solve(a,b)
  return w
```

```python
#Finds rmse, requires two parameters the actual output and the predicted output
def rmse(y,t):
  n=len(y)
  ans=0
  for i in range(0,n):
    ans=ans+(y[i]-t[i])**2
  ans=ans/n
  ans=ans**0.5
  return ans
```

```python
def f(x,w):
  d=len(w)
  ans=0
  v=1
  for i in range(0,d):
    ans=ans+v*w[i]
    v=v*x
  return ans
```

```python
def rms(y,t,lam,w):
    n=len(y)
    ans=0
    for i in range(0,n):
      ans=ans+(y[i]-t[i])**2
    ans=ans/n
    ans=ans**0.5
    sum=0
    for i in w:
      sum=sum+i*i
    ans=ans+(sum*lam*0.5)
    return ans
```
